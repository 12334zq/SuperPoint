```python
# Random generation of cubes
dset = SyntheticDataset()
for i in range(5):
    imgs = []
    labels = []
    for _ in range(3):
        img = dset.generate_background(size)
        pts = dset.draw_cube(img)
        dset.draw_interest_points(img, pts)
        imgs.append(img)
        labels.append(str(i))
    plot_imgs(imgs, ylabel=labels)
```
