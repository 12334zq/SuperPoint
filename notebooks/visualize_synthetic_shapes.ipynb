```python
from superpoint.datasets import synthetic_dataset as dset
from utils import plot_imgs
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
size = (240, 320)
n_show = 3
fcts = [dset.draw_lines, dset.draw_polygon, dset.draw_multiple_polygons, dset.draw_ellipses, 
        dset.draw_star, dset.draw_checkerboard2, dset.draw_cube, dset.gaussian_noise, dset.draw_stripes]
```

```python
for f in fcts:
    r = []
    for _ in range(n_show):
        img = dset.generate_background(size, 30)
        pts = f(img)
        img = dset.draw_interest_points(img, pts)
        r.append(img.copy())
    plot_imgs(r, ylabel=f.__name__,  normalize=True)
```

```python
r = []
for _ in range(1):
    img = dset.generate_background(size, 30)
    pts = dset.draw_checkerboard2(img)
    img = dset.draw_interest_points(img, pts)
    r.append(img.copy())
plot_imgs(r, ylabel=['draw_checkerboard'],  normalize=True, cmap='gray')
```
