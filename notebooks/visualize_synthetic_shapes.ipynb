```python
from superpoint.datasets.synthetic_dataset import SyntheticDataset
from utils import plot_imgs
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
size = (240, 320, 3)
n_show = 3
dset = SyntheticDataset()
fcts = [dset.draw_lines, dset.draw_polygon, dset.draw_multiple_polygons, dset.draw_ellipses, 
        dset.draw_star, dset.draw_checkerboard, dset.draw_cube, dset.gaussian_noise, dset.draw_stripes]
```

```python
for f in fcts:
    r = []
    for _ in range(n_show):
        img = dset.generate_background(size)
        pts = f(img)
        dset.draw_interest_points(img, pts)
        r.append(img.copy())
    plot_imgs(r, ylabel=f.__name__)
```
